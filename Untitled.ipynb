```python
train_size = int(len(data)*.75)
val_size = int(len(data)*0.2)
test_size = int(len(data)*0.05)
```

```python
test_size
```

```python
# ones we have the number of baches for train validation and test, we use the metods take and skip to assing the data
train = data.take(train_size) # take the number of batches for training data
val = data.skip(train_size).take(val_size) # fist skip the datas taked for training and then take the baches for validation
test = data.skip(train_size+val_size).take(test_size) # skoip train and validation datas and take baches for test
```

```python
from tensorflow.keras.models import Sequential # (sequential used for single imput single output) also we can use FUNTIONAL 
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.applications import VGG16
from tensorflow.keras import layers
```

```python

base_model = tf.keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
model = tf.keras.Sequential([
    base_model,
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(6, activation='softmax')
])
```

```python
model = Sequential()
model.add(Conv2D(32, (3,3),1, activation='relu', input_shape=(128,128,3)))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64, (3,3), 1, activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(6, activation='softmax'))
```

```python
tf.keras.losses.SparseCategoricalCrossentropy??
```

```python
model.compile('adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
# adam is the optimizer # for the loss we can configure parameters
```

```python
hist = model.fit(train, epochs=5, validation_data=val, callbacks=[tensorboard_callback])
```

```python
fine_tune_at = 4  # Ajusta este valor según las capas que deseas descongelar
for layer in model.layers[:fine_tune_at]:
    layer.trainable = False
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss=tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
```

```python
# Recompilar el modelo después del fine-tuning
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(train, epochs=10, validation_data=val, callbacks=[tensorboard_callback])
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
from tensorflow.keras.metrics import Precision, Recall, Accuracy
```

```python
pre = Precision()
re = Recall()
acc = Accuracy()
```

```python

y_true =[]
y_pred = [] 
l=0
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    l=l+1
    if l==test_size:
        for i in range(32):
            y_true.append(y[i])
            y_pred.append(yhat[i].argmax())
    else:
        for i in range(32):
            y_true.append(y[i])
            y_pred.append(yhat[i].argmax())
            
    pre.update_state(y, y_pred)
    re.update_state(y, y_pred)
    acc.update_state(y, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
import itertools
cm = confusion_matrix(np.transpose(y_true), np.transpose(y_pred))
classes = ['carton', 'vidrio', 'plastico', 'papel', 'metal', 'basura comun']
def plot_confusion_matrix(cm, classes):
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Etiqueta Verdadera')
    plt.xlabel('Predicción')

# Graficar la matriz de confusión
plot_confusion_matrix(cm, classes)
plt.show()
```

```python
print(f"presition: {pre.result().numpy()}, recall: {re.result().numpy()}, accuracy: {acc.result().numpy()}")
```

```python
# Evaluate the model on test data
for batch in test.as_numpy_iterator(): 
    X_test, y_test = batch
    loss, accuracy =  model.evaluate(X_test, y_test, verbose=2)
print("Test accuracy:", accuracy)
    #print("Test accuracy:", test_acc)
```
